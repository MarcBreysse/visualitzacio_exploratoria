```python
#Exercici 1
#Resumeix gràficament el data set DelayedFlights.csv

#Crea almenys una visualització per:
#Una variable categòrica (UniqueCarrier)
#Una variable numèrica (ArrDelay)
#Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)
#Dues variables numèriques (ArrDelay i DepDelay)
#Tres variables (ArrDelay, DepDelay i UniqueCarrier)
#Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier).
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

Airlines_Delay = pd.read_csv(r'C:\\Users\\mio\\Documents\\IT ACADEMY\\3.5_Pandas\\data\\rawdata\\DelayedFlights.csv')
```

```python
# 'Airlines_Delay_R': 'Airlines_Delay' sense les ultimes columnes
Airlines_Delay_R = Airlines_Delay[['Unnamed: 0', 'Year', 'Month', 'DayofMonth', 'DayOfWeek', 'DepTime',
       'CRSDepTime', 'ArrTime', 'CRSArrTime', 'UniqueCarrier', 'FlightNum',
       'TailNum', 'ActualElapsedTime', 'CRSElapsedTime', 'AirTime', 'ArrDelay',
       'DepDelay', 'Origin', 'Dest', 'Distance']]
Airlines_Delay_R.shape
```

```python
Airlines_Delay_R.columns
```

```python
Airlines_Delay_R.head()
```

```python
#Crea almenys una visualització per:
    #Una variable categòrica (UniqueCarrier)
CarrierVisual=Airlines_Delay_R['UniqueCarrier'].value_counts()
print(CarrierVisual)
```

```python
CarrierVisual.plot.pie(title = "Number of Flight by Carrier")

plt.savefig('NumberFlightByCarrier.jpg')

plt.show()
#altre opció amb 'bar'
    #CarrierVisual.plot.bar(title = "Number of Flight by Carrier", color = "red")
```

```python
#Crea almenys una visualització per:
    #Una variable numèrica (ArrDelay)
Airlines_Delay_R['ArrDelay'].plot(title = "Arrival Delay Distribution",color = "blue")

plt.savefig('ArrivalDelayDistribution.png')

plt.show()
```

```python
#Crea almenys una visualització per:
    #Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)
Airlines_Delay_R.plot(kind='scatter',x='UniqueCarrier',y='ArrDelay',title = "Arrival Delays Per Carrier",color='green')

plt.savefig('ArrDelayPerCarrier.png')

plt.show()
```

```python
#Crea almenys una visualització per:
    #Dues variables numèriques (ArrDelay i DepDelay)
        #amb plot
Airlines_Delay_R.plot(kind='scatter',x='DepDelay',y='ArrDelay',title = "Arrival Delay Vs Departure Delay",color='black',s=2)

plt.savefig('ArrDelayVSDepDelay.png')

plt.show()
```

```python
         #amb seaborn
sns.set()
sns.scatterplot(Airlines_Delay_R['DepDelay'], Airlines_Delay_R['ArrDelay'])

plt.savefig('ArrDelayVSDepDelay_sns.png')
```

```python
#Crea almenys una visualització per:
    #Tres variables (ArrDelay, DepDelay i UniqueCarrier)
ax = plt.gca()
Airlines_Delay_R.plot(kind='scatter',x='UniqueCarrier',y='DepDelay', ax=ax)
Airlines_Delay_R.plot(kind='scatter',x='UniqueCarrier',y='ArrDelay', title = "Arrival Delay & Departure Delay per Carrier", color='red', ax=ax)

plt.savefig('ArrDelay&DepDelayPerCarrier.png')

plt.show() 
```

```python
#Crea almenys una visualització per:
    #Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier).
#sns.pairplot(Airlines_Delay)
#ver https://todoia.es/guia-completa-de-visualizacion-de-datos-con-python/
ax = plt.gca()
Airlines_Delay_R.plot(kind='scatter',x='UniqueCarrier',y='DepDelay', ax=ax)
Airlines_Delay_R.plot(kind='scatter',x='UniqueCarrier',y='ArrDelay', color='red', ax=ax)
Airlines_Delay_R.plot(kind='scatter',x='UniqueCarrier',y='AirTime', title = "Arrival Delay & Departure Delay per Carrier", color='green', ax=ax)
plt.savefig('ArrDelay&DepDelay&AirTimePerCarrier.png')

plt.show() 
```
