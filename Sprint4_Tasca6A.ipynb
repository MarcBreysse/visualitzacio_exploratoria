```python
# graficar del dataframe 'Dummies From AK tutorial'
dummies = pd.read_csv('C:\\Users\\mio\\Documents\\IT ACADEMY\\4.6_Visualització_gràfica\code\\finaldata\\dummies.csv', sep = ',', encoding = 'utf-8')
dummies.head()
```

```python
dummiesT = dummies.T # or df1.transpose()
dummiesT.tail()
```

```python
dummiesT['sum_genero'] = dummiesT.sum(axis=1)
dummiesT.head()
```

```python
dummiesT.plot(y='sum_genero',kind = 'barh',figsize=(7, 7))
plt.xlabel("Cantidad de peliculas")
plt.ylabel("Genero")
plt.title("SUMA DE GENEROS")
```

<span style=color:green><font size="3">Graficar 10 PRIMEROS GENEROS ".</span>

```python
dummiesT.iloc[0:10].plot.pie(y='sum_genero',autopct='%1.1f%%', pctdistance=0.8, startangle=170, rotatelabels = 270, figsize=(8, 8))
plt.xlabel("Cantidad de peliculas")
plt.title("10 PRIMEROS GENEROS")
```

<span style=color:green><font size="3">Graficar SUMA DE PELICULAS POR AÑO y EVOLUCIÓN DE GENEROS CON EL TIEMPO (con archivo Dummies completo)".</span>

```python
# graficar del dataframe 'Dummies From AK tutorial'
movies_dummies = pd.read_csv('C:\\Users\\mio\\Documents\\IT ACADEMY\\4.6_Visualització_gràfica\code\\finaldata\\movies_dummies.csv', sep = ',', encoding = 'utf-8')
```

```python
movies_dummies.tail(10)
```

```python
movies_dummies['año'] = movies_dummies['titulo'].str.strip().str[-5:-1]
# tambien valido.....movies_dummies['año'] = [x.strip()[-5:-1] for x in movies_dummies['titulo']]
movies_dummies['titulo_solo'] = movies_dummies['titulo'].str.strip().str[:-6]
movies_dummies.drop(columns =["titulo"], inplace = True) 
movies_dummies.head()
```

```python
movies_dummies_year = movies_dummies['año'].value_counts()
movies_dummies_year.head()
```

<span style=color:green><font size="3">Grafico '10 MEJORES AÑOS'.</span>

```python
movies_dummies_year.iloc[0:10].plot.barh()
plt.xlabel("Cantidad de peliculas")
plt.ylabel("Año de estreno")
plt.title("MEJORES AÑOS DE PRODUCCIÓN DE PELICULAS")
#df.iloc[:,0:3].plot.bar(figsize = size, ax = ax3)
```

<span style=color:green><font size="3">Grafico '50 MEJORES AÑOS'.</span>

```python
plt.subplot(1, 3, 1)
movies_dummies_year.iloc[0:17].plot.barh(figsize=(10, 4))
plt.ylabel("Año de estreno")
plt.subplot(1, 3, 2)
movies_dummies_year.iloc[17:34].plot.barh()
plt.subplot(1, 3, 3)
plt.xlabel("Cantidad de peliculas")
movies_dummies_year.iloc[34:50].plot.barh()
plt.suptitle("50 MEJORES AÑOS DE PRODUCCIÓN")
```

<span style=color:green><font size="3">Grafico 'MEJORES AÑOS DE PRODUCCIÓN DE PELICULAS (Pie Plot)'.</span>

```python
movies_dummies_year.iloc[0:10].plot.pie(title = "MEJORES AÑOS DE PRODUCCIÓN DE PELICULAS")
plt.show()
```

<span style=color:green><font size="3">Grafico EVOLUCIÓN DE GENEROS CON EL TIEMPO.</span>

```python
movies_dummies.head(7)
```

```python
movies_dummies_per_ano = movies_dummies.iloc[:,2:21]     ## selección of 'generos' y 'año' 
movies_dummies_per_ano.head(7)
# selección of columns with 'float' values.....movies_dummies.select_dtypes(include=['float']).head()
```

```python
movies_dummies_per_ano_grouped = movies_dummies_per_ano.groupby('año').sum()
movies_dummies_per_ano_grouped.tail()
```

```python
movies_dummies_per_ano_grouped[['Genero_Drama','Genero_Comedy','Genero_Action']].plot(title = "EVOLUCIÓN DE LOS 3 GENEROS MAS POPULARES CON EL TIEMPO")
```

```python
sns.catplot(y='Genero_Drama', kind='box',
                 data=movies_dummies_per_ano_grouped);
```
